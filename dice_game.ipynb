```python
def dicegame():
    
    print('Welcome to Dice Game')
    
    while True:
        
        from random import randint
        from IPython.display import clear_output
        
        luck = input("press 'space and enter' to see which player will go first")
        while luck != ' ':
            luck = input("press space and enter to see which player will go first")
        clear_output()
        
        player1_value = []
        player2_value = []
        
        choose_players = randint(1,2)
        if choose_players == 1:
            turn = 'player1'
        else:
            turn = 'player2'
        
        print(f"{turn} will play first")
       
        game_on = input('Do you want to play this game? y or n ').upper()
        clear_output()
        if game_on == 'Y':
            game_on = True
        else:
            game_on = False
            print("üëã üëã")
            break
            
        while True:
            try:
                value = int(input('Choose a target value: '))
                break
            except ValueError:
                print("choose only whole numbers")
        
        print("Let's start!\n")
        
        while game_on == True:
        
            if turn == 'player1':
                choose = input("Player1: Hit 'space and enter' to throw üé≤üé≤")
                if choose == ' ':
                    x = randint(1,6)
                    print(f"     {x} on üé≤     ")
                    player1_value.append(x)
                
                    total = 0
                    for num in player1_value:
                        total = total + num
                    print(f"\ntotal = {total}\n")
                    
            
                    if total >= value:
                        print('player1 üèÜ')
                        game_on = False
                    else: 
                        turn != 'player1'
                        turn = 'player2'        
                else: 
                    turn != 'player1'
                    turn = 'player2'
                                
            if turn == 'player2':
                choose = input("Player2: Hit 'space and enter' to throw üé≤üé≤")
                if choose == ' ':
                    y = randint(1,6)
                    print(f"     {y} on üé≤      ")
                    player2_value.append(y)
                
                    total = 0
                    for num in player2_value:
                        total = total + num
                    print(f"\ntotal = {total}\n")
                    
                    if total >= value:
                        print('player2 üèÜ')
                        game_on = False
                    else:
                        turn != 'player2'
                        turn = 'player1'
                else:
                    turn != 'player2'
                    turn = 'player1'
            
        if not game_on:
            replay = input("Do you want to replay this game? y or n ").upper()
    
        if replay == 'Y':
            game_on = True
        else: 
            print("Thanks for playing")
            break
        
```
