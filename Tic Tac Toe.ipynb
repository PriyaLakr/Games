```python
def tictactoe():
    
    while True:
        Times = []    
        player1 = True
        player2 = False
        Turns = [1,2,3,4,5,6,7,8,9]
        
        
        from IPython.display import clear_output
        
        x = input(" Player1: Choose symbol from X or O " ).upper()
        clear_output()
        if x == 'X':
            y = 'O'
        elif x == 'O':
            y = 'X'
        print(f'Player1 symbol is {x} and Player2 symbol is {y}')
            
        
        A = [' # ',' ',' ',' ',' ',' ',' ',' ',' ',' ','|']
        e = "-----------"   
        print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
    
        game_on = input("Do you want to play this game? y or n ").upper()
        clear_output()
        
        if game_on == 'Y':
            game_on = True
        else:
            game_on = False
            break
    
        while game_on == True:
            
            if player1 == True and player2 == False: 
                
                position1 = 0
                while position1 not in Turns or A[position1] != ' ':
                    position1 = int(input('Player1: choose a position from the board (1-9)'))
        
                A[position1] = x  
                Times.append(position1)
                print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
                clear_output()
        
                if (A[1] == x and A[2] == x and A[3] == x) or (A[7] == x and A[8] == x and A[9] == x) or (A[1] == x and A[4] == x and A[7] == x) or (A[2] == x and A[5] == x and A[8] == x) or (A[1] == x and A[5] == x and A[9] == x) or (A[4] == x and A[5] == x and A[6] == x) or (A[3] == x and A[6] == x and A[9] == x) or (A[3] == x and A[5] == x and A[7] == x):
                    print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
                    print('Yay! Player1 wins!')
                    game_on = False
                elif set(Times) == set(Turns):
                    print("It's a draw!")
                    game_on = False
                else:
                    print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}') 
                    player2 = True
                    player1 = False
                
        
            if player1 == False and player2 == True:
        
                position2 = 0
                while position2 not in Turns or A[position2] != ' ': #while the position you are selecting is not empty(i.e. filled; A[position1] != ' '), keep on asking new position, till you encounter an empty position A[position1] == ' '
                    position2 = int(input('Player2: choose a position from the board (1-9)'))
                
                A[position2] = y  
                Times.append(position2)
                print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
                clear_output()
            
                if (A[1] == A[2] == A[3] == y) or (A[7] == A[8] == A[9] == y) or (A[1] == A[4] == A[7] == y) or (A[2] == A[5] == A[8] == y) or (A[1] == A[5] == A[9] == y) or (A[4] == A[5] == A[6] == y) or (A[3] == A[6] == A[9] == y) or (A[3] == A[5] == A[7] == y):
                    print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
                    print('Yay! Player2 wins!')
                    game_on = False
                elif set(Times) == set(Turns):
                    print('Its a draw!')
                    game_on = False
                else:
                    print(f' {A[1]} {A[10]} {A[2]} {A[10]} {A[3]} \n{e} \n {A[4]} {A[10]} {A[5]} {A[10]} {A[6]} \n{e} \n {A[7]} {A[10]} {A[8]} {A[10]} {A[9]}')   
                    player2 = False
                    player1 = True
                
                 
        if not game_on:
            replay = input("Do you want to replay this game? y or n ").upper()
    
        if replay == 'N':
            break
```
